This exercise explores the `Seaborn` package, a Python library for making statistical plots using `matplotlib` and with `pandas` data structures.

Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots.

## Load an example dataset

```python
tips = sns.load_dataset("tips")
```

```python
# Import seaborn package
import seaborn as sns
```

## View data structure

```python
tips.head(n=10)
```

## Create a visualization
Draw a scatterplot by including a linear regression model using `lmplot()`.

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")

#sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="sex")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", hue="time")
```
